## Robinson Operator

This operator is also known as direction mask.
In this operator we take one mask and rotate it in all
the 8 compass major directions to calculate edges of each direction.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im = cv.imread('assets/cameraman.png', 0)
```

```python
plt.imshow(im, cmap=plt.cm.gray)
plt.title('Original Image')
```

```python
im.shape
```

```python
k = np.array([
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
])
e1 = cv.filter2D(im, -1, k)

k = np.array([
    [0, -1, -2],
    [-1, 0, 1],
    [2, 1, 0]
])
e2 = cv.filter2D(im, -1, k)

k = np.array([
    [1, 0, -1],
    [2, 0, -2],
    [1, 0, -1]
])
e3 = cv.filter2D(im, -1, k)

k = np.array([
    [2, 1, 0],
    [1, 0, -1],
    [0, -1, -2]
])
e4 = cv.filter2D(im, -1, k)

k = np.array([
    [1, 2, 1],
    [0, 0, 0],
    [-1, -2, -1]
])
e5 = cv.filter2D(im, -1, k)

k = np.array([
    [0, 1, 2],
    [-1, 0, 1],
    [-2, -1, 0]
])
e6 = cv.filter2D(im, -1, k)

k = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])
e7 = cv.filter2D(im, -1, k)

k = np.array([
    [-2, -1, 0],
    [-1, 0, 1],
    [0, 1, 2]
])
e8 = cv.filter2D(im, -1, k)
```

```python
plt.imshow(e1, cmap=plt.cm.gray)
plt.title('North')
plt.show()

plt.imshow(e2, cmap=plt.cm.gray)
plt.title('NorthWest')
plt.show()

plt.imshow(e3, cmap=plt.cm.gray)
plt.title('West')
plt.show()

plt.imshow(e4, cmap=plt.cm.gray)
plt.title('SouthWest')
plt.show()

plt.imshow(e5, cmap=plt.cm.gray)
plt.title('South')
plt.show()

plt.imshow(e6, cmap=plt.cm.gray)
plt.title('SouthEast')
plt.show()

plt.imshow(e7, cmap=plt.cm.gray)
plt.title('East')
plt.show()

plt.imshow(e8, cmap=plt.cm.gray)
plt.title('NorthEast')
plt.show()
```
